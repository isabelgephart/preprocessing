# This notebook is the first step in preprocessing an fMRI dataset

## It creates folder structure for a new project

# Folder structure

## SELECT PROJECT NAME:

```python
PRJ = '28andMe'
```

```python
import os.path as osp
import os
```

```python
PRJ_DIR = osp.join('/project/mdrosenberg/IG', PRJ)
```

```python
#PRJ_DIR
if not osp.exists(PRJ_DIR):
    os.makedirs(PRJ_DIR)
    print('Created directory: ', PRJ_DIR)
else:
    print('Directory exists: ', PRJ_DIR)

#CODE_DIR
c = osp.join(PRJ_DIR, 'code')
if not osp.exists(c):
    os.makedirs(c)
    print('Created directory: ', c)
else:
    print('Directory exists: ', c)

#notebooks
n = osp.join(c,'notebooks')
if not osp.exists(n):
    os.makedirs(n)
    print('Created directory: ', n)
else:
    print('Directory exists: ', n)

#bash
b = osp.join(c,'bash')
if not osp.exists(b):
    os.makedirs(b)
    print('Created directory: ', b)
else:
    print('Directory exists: ', b)

#python
p = osp.join(c,'python')
if not osp.exists(p):
    os.makedirs(p)
    print('Created directory: ', p)
else:
    print('Directory exists: ', p)

#SBATCH
s = osp.join(c,'SBATCH')
if not osp.exists(s):
    os.makedirs(s)
    print('Created directory: ', s)
else:
    print('Directory exists: ', s)

#derivatives
d = osp.join(PRJ_DIR,'derivatives')
if not osp.exists(d):
    os.makedirs(d)
    print('Created directory: ', d)
else:
    print('Directory exists: ', d)
    
#fmriprep
fprep = osp.join(PRJ_DIR,'derivatives','fmriprep')
if not osp.exists(fprep):
    os.makedirs(fprep)
    print('Created directory: ', fprep)
else:
    print('Directory exists: ', fprep)
    
#preprocessed
p = osp.join(PRJ_DIR,'derivatives','preprocessed')
if not osp.exists(p):
    os.makedirs(p)
    print('Created directory: ', p)
else:
    print('Directory exists: ', p)

#work
w = osp.join(PRJ_DIR,'derivatives','work')
if not osp.exists(w):
    os.makedirs(w)
    print('Created directory: ', w)
else:
    print('Directory exists: ', w)
    
```

### The setup is now done.
